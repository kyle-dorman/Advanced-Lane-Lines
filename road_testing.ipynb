```python
%matplotlib inline

def hist(data, name):
  n, bins, patches = plt.hist(data, bins=50, facecolor='green')

  plt.xlabel(name)
  plt.ylabel('Count')
  plt.title('Count  of ' + name)
  plt.grid(True)
  plt.show()
  
hist(road.slope_diffs, "slope diffs")
hist(np.array(road.fit_diffs)[:,0], "fit_diffs: a")
hist(np.array(road.fit_diffs)[:,1], "fit_diffs: b")
hist(np.array(road.fit_diffs)[:,2], "fit_diffs: c")
hist(road.lane_distance, "lane_distance")
```
